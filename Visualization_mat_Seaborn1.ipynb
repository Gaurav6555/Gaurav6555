##### Visualizing Data with pandas

## Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Gaurav/Downloads/train.csv")
```

```python
df
```

```python
sns.set_style("darkgrid",{'axes.grid' : True})
sns.lmplot(x= 'PassengerId',y='Age',data=df)
plt.show()
```

```python
sns.lmplot(x= 'PassengerId',y='Age',data=df, hue="Sex", palette='Set1')
plt.show()
```

```python
sns.lmplot(x= 'PassengerId',y='Age',data=df, col="Sex")
plt.show()
```

### Univariate - "one variable" data visualization

```python
#strip plot

sns.stripplot(y= 'Age', data=df, jitter=True)
plt.ylabel('Age ($)')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x= 'Pclass', y='PassengerId', data=df,)
plt.ylabel('Pclass ($)')
plt.show()
```

```python
sns.stripplot(x= 'Pclass', y='PassengerId', data=df,size=4,jitter=False)
plt.ylabel('Pclass ($)')
plt.show()
```

```python
#Swarm plot
sns.swarmplot(x= 'Pclass', y='PassengerId', data=df)
plt.ylabel('Pclass ($)')
plt.show()
```

```python
sns.swarmplot(x= 'Pclass', y='PassengerId', data=df,hue='Sex',)
plt.xlabel('Pclass ($)')
plt.show()
```

```python
#Changing orientation
sns.swarmplot(x= 'Pclass', y='PassengerId', data=df,hue='Sex',orient='h')
plt.xlabel('Pclass ($)')
plt.show()
```
